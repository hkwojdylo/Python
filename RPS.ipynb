```python
class Player:
    def __init__(self, name):
        self.name = name
        self.score = 0
        self.throws = {}
        
    def __repr__(self):
        curr = self.name + " has a score of " + str(self.score) + " after playing "
        print(curr)
        print(self.throws)
        return ""
    
    def throw(self, x):
        if x in self.throws:
            self.throws[x] +=1
        else:
            self.throws[x] = 1
            
        
               
    
```
