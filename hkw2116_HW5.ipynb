# define wrLines() and wrChars() functions
- wrLines writes the contents of 'tobeLines' to a file, using the fd.writelines() method. 
you must add a '\n' explicitly to each line. 
- wrChars writes the contents of 'tobe', one character at a time, using fd.write() method
- after the file write is complete, both functions should get the file byte size,
using tp.stat(), and return it.
- note the file length on windows will be a little larger, because windows inserts a carridge return(\r) before each newline(\n)

````python
def wrLines(x, lines):
    with open(x, 'w') as fd:
        fd.writelines(i + '\n' for i in lines)
        
    stat = tp.stat()
    return stat.st_size
````

````python
def wrChars(x, lines):
    with open(x, 'w') as fd:
        for i in lines:
            fd.write(i)
        
    stat = tp.stat()
    return stat.st_size
````

# define rdAll()
- read the tp file in one fd.read() call (eagar approach)
- return the number of characters read, and index of index of search string, or -1 if not found


````python
def rdAll(x, fi):
    with open(x, 'r') as fd:
        ham = fd.read()
        
    try:
        ind = ham.index(fi)
        return (len(ham), ind)
    except Exception as e:
        ind = -1
        return (len(ham), ind)
    
````

# define rdLines()
- read the tp file by using fd.readlines() (eager approach)
- return the number of chars searched, and the line number where the search string was found, otherwise -1

````python
def rdLines( x, fi):
    index = -1
    count = 0
    with open(x, 'r') as fd:
        ham = fd.readlines()
        for i in range(len(ham)):
            count += len(ham[i])
            if fi in ham[i]:
                index = i
                break
            
    return (count, index)
````

# define rdLine()
- read the tp file by using fd.readline() (lazy approach)
- return the number of chars read, and the line number where the search string was found, otherwise -1

````python
def rdLine(x, fi):
    index = -1
    count = 0
    with open(x, 'r') as fd:
        while True:
            line = fd.readline()
            index = index+1
            count += len(line)
            if fi in line:
                return (count, index)
    
            if line == '':
                break;
        
        return (count, -1)
````

````python
[ (s, rdLine(tp, s)) for s in ('foobar', 'not', 'perchance', 'insolence')]
````

# define rdChar
- read the tp file one character at a time using fd.read(1) (lazy approach)
- fd.read(1) will return an empty string when the file is exhausted
- read the min # of chars to find the search string
- return (# chars read, position of string)


````python
def rdChar(x, fi):
    ham = ""
    
    with open(x, 'r') as fd:
        while True:
            let = fd.read(1)
            ham = ham + let
            
            index = ham.find(fi)
            if index != -1:
                return (len(ham), index)
            
            if let == '':
                break;
            
        return (len(ham), index)
````

````python
[ (s, rdChar(tp, s)) for s in ('foobar', 'not', 'perchance', 'insolence')]
````

# define rdMap
- memory map the tp file (lazy approach)
- return position of search string


````python
import mmap

def rdMap(x, fi):
    fp = open(x, 'r+')
    fn = fp.fileno()
    map = mmap.mmap(fn,0)
    
    map.seek(0)
    return(map.find(fi.encode()))
````

````python
[ (s, rdMap(tp, s)) for s in ('foobar', 'not', 'perchance', 'insolence')]
````

````python
def hamlet():
    lncnt = 0
    speechcnt = 0
    speeches = {}
    with open('hamlet.html', 'rb') as fd: 
        while True:
            line = fd.readline()
            lncnt = lncnt+1
            
            if b'NAME=speech' in line:
                speechcnt = speechcnt+1
                i = line.find(b'<b>')
                j = line.find(b'</b>')
                key = line[i+3:j]
                
                if key in speeches:
                    speeches[key] = speeches[key]+1
                else:
                    speeches[key] = 1
            
            if line == b'':
                break;
                
        return (lncnt, speechcnt, speeches)
````

````python
from datetime import datetime
from datetime import date
import pandas as pd
````

````python
# with the arg defaults shown, this will 
# calculate the money accumulated after 
# 31 days, with weekly deposits of $10, 
# at a 10% interest rate compounded
# on each business day

daysofWeek = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday' }

def savings(start = '2018-03-01', end = '2018-03-31',\
            depositFreq='W', depositAmount=10., 
            compoundFreq='B', interestRate=.1 ):
    
    dates = pd.date_range(start, end).to_series()
    money= pd.DataFrame(index=dates).fillna(0.0)
    
    s = start.split('-')
    d = date(int(s[0]), int(s[1]), int(s[2])).weekday()
    dayofweek = []
    for i in range(len(dates)):
        dayofweek.append(daysofWeek[(i+d)%7])

    money['day'] = dayofweek
    
    deposits = pd.Series(depositAmount, index = pd.date_range(start, end, freq = depositFreq))
    
    compound = pd.Series(interestRate, index = pd.date_range(start, end, freq = compoundFreq))
    money['deposits'] = deposits
    money['compound'] = compound

    money = money.fillna(0)
    
    balser = pd.Series(0.0, index = dates)
    intser = pd.Series(0.0, index = dates)
    
    balance = 0.0
    interest = 0.0
    
    
    for j,t in enumerate(money.itertuples()):
        # pull out row values
        td = t.deposits
        ti = t.compound
        # calculate interest and new balance
        balance += td
        interest = balance * ti
        balance += interest
        # update the series objects
        balser[j] = balance
        intser[j] = interest
    
    
        
    money['interest'] = intser
    money['balance']= balser
    
    return money
    
    pass
    
````
